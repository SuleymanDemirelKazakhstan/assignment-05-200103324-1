```python
import pandas as pd
data = pd.read_csv("IMDB.csv")
data = data.head(10)
y = data['sentiment'].replace({'positive': 1, 'negative': 0})
data
```

```python
inputMeans=[]
for index, row in data.iterrows():
    inputMeans.append([])
    rowText = nlp(row['review'])
    for token in rowText:
        lineToken= nlp(token.text)
        vectorTokenized = lineToken[0].vector
        meanOfVector = vectorTokenized.mean()
        inputMeans[index].append(meanOfVector)
```

```python
x= pd.DataFrame(inputMeans)
x
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,shuffle=True,random_state=4)
```

```python

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

logreg = LogisticRegression(random_state=16)


logreg.fit(X_train, y_train)



y_pred = logreg.predict(X_test)
score =accuracy_score(y_test,y_pred)
score
```
